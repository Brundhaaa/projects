```python
!apt-get install -y fluidsynth
!pip install music21 midi2audio
from music21 import stream, note, midi
import random

def generate_ai_melody(length=16):
    s = stream.Stream()
    pitch = 60
    for _ in range(length):
        interval = random.choice([-2, -1, 0, 1, 2])
        pitch = max(48, min(72, pitch + interval))
        duration = random.choice([0.25, 0.5, 1.0])
        n = note.Note(pitch)
        n.quarterLength = duration
        s.append(n)
    return s

melody = generate_ai_melody()
mf = midi.translate.streamToMidiFile(melody)
midi_path = "/content/ai_music.mid"
mf.open(midi_path, 'wb')
mf.write()
mf.close()
from midi2audio import FluidSynth

# Convert MIDI to WAV
fs = FluidSynth()
audio_path = "/content/ai_music.wav"
fs.midi_to_audio(midi_path, audio_path)
from IPython.display import Audio
Audio(audio_path)
```
